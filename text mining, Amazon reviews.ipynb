Here we select 25 reviews of samsung galaxy s20 fe and we are performing emotion mining.

```python
import pandas as pd
import numpy as np
```

```python
with open("C:\\ExcelR course\\Data science new\\assignments\\Text Mining\\reviews.txt","r") as input:
    S20_Fe_reviews = input.read()
type(S20_Fe_reviews)
```

```python
S20_Fe_reviews
```

```python
import re
```

```python
s20_rev_string = S20_Fe_reviews
s20_rev_string = re.sub("[^A-Za-z" "]+"," ",s20_rev_string).lower()
s20_rev_string = re.sub("[0-9" "]+"," ",s20_rev_string) 
```

```python
s20_rev_string
```

```python
s20_rev_words = s20_rev_string.split(" ")
```

```python
s20_rev_words
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
s20_rev_words = [w for w in s20_rev_words if not w in stop_words]
```

```python
s20_rev_string = " ".join(s20_rev_words)
```

```python
s20_rev_string
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud_s20 = WordCloud(
                        background_color="white",
                       ).generate(s20_rev_string)

plt.imshow(wordcloud_s20)
plt.axis("off")
plt.show()
```

Positive words

```python
with open("C:\\ExcelR course\\Data science new\\assignments\\Text Mining\\positive-words.txt","r") as pos:
    poswords = pos.read().split("\n")
```

Negative words

```python
with open("C:\\ExcelR course\\Data science new\\assignments\\Text Mining\\negative-words.txt","r") as neg:
    negwords = neg.read().split("\n")
```

Positive wordcloud

```python
ip_pos_in_pos = " ".join([w for w in s20_rev_words if w in poswords])

wordcloud_pos_in_pos = WordCloud(
                       background_color="white",
                      ).generate(ip_pos_in_pos)
plt.imshow(wordcloud_pos_in_pos)
plt.axis("off")

plt.show()
```

Negative wordcloud

```python
ip_neg_in_neg = " ".join([w for w in s20_rev_words if w in negwords])

wordcloud_neg_in_neg = WordCloud(
                       background_color="black",
                      ).generate(ip_neg_in_neg)
plt.imshow(wordcloud_neg_in_neg)
plt.axis("off")

plt.show()
```
